### **Código de la Red Neuronal**

En esta sección se encuentra el codigo de entrenamiento de la red neuronal que fue configurada anteriormente. El codigo se encuentra debidamente comentado, resaltando las partes importantes en el mismo.

```python
# IMPORTAR LIBRERIAS

# Libreria de procesamiento numerico
import numpy as np
# Libreria que añade y exponde las funcionalidades de numphy
import scipy as sc
# Libreria para graficos
import matplotlib.pyplot as plt
# Libreria para controlar tiempo de ejecucion
import time
#Libreria para control de pantalla
from IPython.display import clear_output
```

```python
# IMPORTAR DATOS

def import_file(file_name,cols):
    data = np.loadtxt(file_name, delimiter=',', skiprows=1, usecols=cols, dtype=float)
    return data
    
X = import_file(X_file_name, [0,1])
Y = import_file(X_file_name, [2])
W = import_file(W_file_name, [0,1])
T = import_file(T_file_name, [0])

# Arreglar indices de vectores y matrices
Y = Y[:, np.newaxis]
W = W[:, np.newaxis]




```

```python
# CLASE DE LA CAPA DE LA RED

class neural_layer():
    # Se inicializa la claserecibiendo los parametros:
    # n_conn: Numero de conexiones
    # n_neur: Numero de neuronas
    # act_f: Función de activación
    def __init__ (self, n_conn, n_neur, act_f):
        
        #Se asigna a la capa creada una propiedad la funcion de activacion que ejecuta
        self.act_f = act_f
        
        #Inicializar los parametros de la capa
        self.W = W
        self.T = T
```

```python
# FUNCIONES DE ACTIVACION

step = lambda x: x>=0

#PRUEBA DE LA FUNCION DE ACTIVACION ESCALON
#_x = np.linspace(-5,5,100)
#plt.plot(_x, step(_x))
```

```python
# DISEÑAR CAPAS DE RED NEURONAL

# Se define la funcion que crea las capas de la red
def create_nn(topology, act_f):
    
    # Vector que contiene las capas que conforman la red
    nn = []
    
    # Recorre la topologia de la red neuronal y crea las capas
    for l, layer in enumerate(topology[:-1]):
        
        nn.append(neural_layer(topology[l], topology[l+1], act_f))
        
    return nn
```

```python
# DETERMINAR TOPOLOGIA
topology = [len(X[0]), len(Y[0])]

#ENTRENAMIENTO

# Definir la funcion de coste, la funcion de calculo de los errores lineales
cost_f = (lambda Yp, Yr: np.mean((Yp - Yr) ** 2),
          lambda Yp, Yr: (Yp - Yr)
         )

# Se define la funcion en donde se hara el entrenamiento
def train(neural_net, X, Y, cost_f, learning_ratio=0.5, train=True):
    
    # Se define el vector donde se guardara el valor de la suma ponderada y su valor de activacion, por capa
    out = [(None, X)]
    
    # FORWARD PASS
    for l, layer in enumerate(neural_net):
        
        # Suma ponderada de la capa l; Sj en el documento
        z = out[-1][1] @ neural_net[l].W + neural_net[l].T

        # Valor de activacion o de salida de la suma ponderada (z) en la capa l; YRi en el documento
        a = neural_net[l].act_f(z)
        
        # Guardar la suma ponderada y el valor de activacion de la capa l
        out.append((z,a))
        
    if train:
        
        # BACKWARD PASS
        # delta en libros, error lineal en la clase
        err_lineal = []
        
        for l in reversed(range(0, len(neural_net))):
            
            z = out[l+1][0]
            a = out[l+1][1]
            
            err_lineal.insert(0, cost_f[1](a, Y))
            
            #print(neural_net[l].W)
            
            # ACTUALIZACION DE PESOS Y UMBRAL
            neural_net[l].T = neural_net[l].T - np.mean(err_lineal[0], axis=0, keepdims=True) * learning_ratio
            neural_net[l].W = neural_net[l].W - out[l][1].T @ err_lineal[0] * learning_ratio
    return out[-1][1]
```

```python
def interval_gen(_x_):

  # Intervalo del eje x
  x00=(_x_[:,0][_x_[:,0]!=0].min())
  x01=(_x_[:,0][_x_[:,0]!=0].max())
  global x0_interval
  x0_interval = [x00-1, x01+1]

  # Intervalo del eje y
  y00=(_x_[:,1][_x_[:,1]!=0].min())
  y01=(_x_[:,1][_x_[:,1]!=0].max())
  global y0_interval
  y0_interval = [y00-1, y01+1]
```

```python
# Generar malla para dibujo
def graphic_gen(x_parameter, y_parameter, neural_n):

  _x0 = np.linspace(x0_interval[0], x0_interval[1], res)
  _y0 = np.linspace(y0_interval[0], y0_interval[1], res)

  _Y = np.zeros((res, res))

  for i0, x0 in enumerate(_x0):
      for i1, y0 in enumerate(_y0):
          _Y[i0, i1] = train(neural_n, np.array([[x0, y0]]), y_parameter, cost_f, train=False)[0][0]

  plt.pcolormesh(_x0, _y0, _Y, cmap="coolwarm")
  plt.axis("equal")

  plt.scatter(x_parameter[y_parameter[:, 0] == 0, 0], x_parameter[y_parameter[:, 0] == 0, 1], c="skyblue")
  plt.scatter(x_parameter[y_parameter[:, 0] == 1, 0], x_parameter[y_parameter[:, 0] == 1, 1], c="salmon")
  clear_output(wait=True)
  plt.show()
```

```python


def execute(x_parameter, y_parameter, neural_n):

  # Vector donde se almacenan los errores de un patron
  global err_iterations
  err_iterations = []

  interval_gen(x_parameter)

  for i in range(iterations):
      
      # Se entrena a la red
      Y_iteration = train(neural_n, x_parameter, y_parameter, cost_f, learning_ratio)
      
      if i % interval_show == 0:        
          
          # Añadir error del patron
          err_iterations.append(cost_f[0](Y_iteration, y_parameter))
          

          # Generar malla para dibujo
          graphic_gen(x_parameter,y_parameter, neural_n)


          plt.plot(range(len(err_iterations)), err_iterations)
          plt.show()
          time.sleep(1)

          if err_iterations[-1] <= err_max:
              return neural_n
```
